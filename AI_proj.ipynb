```python
import numpy as np
import pandas as pd
import difflib # we use this library to get the closest match of the movie from the available data set for the name entered by the user .
from sklearn.feature_extraction.text import TfidfVectorizer # here TEXTUAL data is converted into numerical data
from sklearn.metrics.pairwise import cosine_similarity # gives the highest similarity score,based on this score we recommend movies.
```

Data Collection and Pre-Processing

```python
# loading the data from the csv file to a pandas dataframe
movies_data = pd.read_csv('/content/drive/MyDrive/AI proj data set/movies.csv')
```

```python
# printing the first 5 rows of the dataframe
movies_data.head()
```

```python
# To know how many number of rows and columns are present in the data frame
movies_data.shape
```

```python
# selecting the relevant features for recommendation

req_features = ['genres','keywords','tagline','cast','director','overview']
print(req_features)
```

```python
# replacing the null valuess with null string

for feature in req_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 6 selected features to the total_combined_features variable.

total_combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '+movies_data['overview']
```

```python
# printing all the values present in total_combined_features
print(total_combined_features)
```

```python
# converting the text data to feature vectors
vector = TfidfVectorizer()
```

```python
# fitting the vector data into a feature vector under the variable f_vector
f_vector = vector.fit_transform(total_combined_features)
```
